```python
import seaborn as sns
import pandas as pd
```

```python
#Bar plot
data=pd.DataFrame({"class":["A","B","C","D","E"],
                   "students":[90,46,35,67,55]
                  })
sns.barplot(x="class",y="students",data=data)
plt.title("Bar plot")
plt.show()
```

```python
#count plot
sns.countplot(x="class",data=data)
plt.title("Count plot")
plt.show()
```

```python
import numpy as np
```

```python
# Histogram (It works on univariate dataset )

data = np.random.randn(100) 
plt.hist(data, bins=10) 
plt.title("Histogram") 
plt.show()
```

This graph shows data is non uniformly distributed.

```python
#Histogram                 #or using seaborn
sns.histplot(data,bins=10)
plt.title("Histogram with KDE")
plt.show()
```

```python
# KDE plot

sns.kdeplot(data, shade=True) 
plt.title("KDE Plot") 
plt.show() 
```

```python
#or using seaborn
sns.histplot(data, bins=10, kde=True) 
plt.title("Histogram with KDE") 
plt.show()
```

```python
# Boxplot (To detect outliers)

sns.boxplot(y=data) 
plt.title("Box Plot") 
plt.show() 
```

```python
#violin plot (boxplot+KDE)

sns.violinplot(y=data) 
plt.title("violin Plot") 
plt.show() 
```

```python
#Scatter plot

x=[1,2,3,4,5]
y=[5,4,7,10,8]
plt.scatter(x,y)
plt.title("scatter plot")
plt.show()
```

```python
# Loading dataset
df = sns.load_dataset('iris') 
df
```

```python
#pair plot
sns.pairplot(df,hue='species')
plt.title("pair plot")
plt.show()
```

```python
# Drop Categorical column

df_numeric = df.drop(columns='species')     #  recommended
# or
df_numeric = df.drop('species', axis=1)     #  alternate form
correlation_matrix = df_numeric.corr()
correlation_matrix
```

```python
df=pd.read_csv("data.csv")
df
```

```python
#Heatmap plot
corr=df.corr()
sns.heatmap(correlation_matrix,annot=True,cmap='coolwarm')
plt.title("Correlation Heatmap")
```

```python
#pie chart

labels=['A','B','C']
sizes=[30,50,20]
plt.pie(sizes,labels=labels,autopct='%1.1f%%',startangle=0)
plt.title("pie chart")
plt.show()
```

```python
# Swarm plot

sns.swarmplot(x='species', y='sepal_length', data=df) 
plt.title("Swarm Plot") 
plt.show() 
```

```python
# Strip plot

sns.stripplot(x='species', y='sepal_length', data=df, jitter=True) 
plt.title("Strip Plot") 
plt.show()
```

```python
# Joint plot  (Scatter+Histogram)

sns.jointplot(x='sepal_length', y='petal_length', data=df, kind='scatter') 
plt.show()
```
